```python
import pandas as pd
import re

df = pd.read_csv("/home/bhikkhu/Bodhirasa/Dropbox/dpd/csvs/dpd-full.csv", sep = "\t", dtype=str)
df = df.fillna("") 
df_length = df.shape[0]
```

```python
with open("output/simple_js.js", "w") as simple_js, open("output/tpp.js", "w") as tpp_js, open("output/html.html", "w") as html_file:

	tpp_js.write("var pv1 = {\n")
	simple_js.write("var pv1 = {\n")
	html_file.write(f"<html>\n")
	
	for row in range(df_length):
		pali = df.loc[row, "Pāli1"]
		pali_clean = re.sub(" \d*$", "", pali)
		pos = df.loc[row, "POS"]
		grammar = df.loc[row, "Grammar"]
		neg = df.loc[row, "Neg"]
		verb = df.loc[row, "Verb"]
		trans = df.loc[row, "Trans"]
		case = df.loc[row, "Case"]
		meaning = df.loc[row, "Meaning IN CONTEXT"]
		lit = df.loc[row, "Literal Meaning"]
		buddhadatta = df.loc[row, "Buddhadatta"]
		non_ia = df.loc[row, "Non IA"]
		sk = df.loc[row, "Sanskrit"]
		sk_root = df.loc[row, "Sk Root"]
		sk_root_mn = df.loc[row, "Sk Root Mn"]
		sk_root_cl = df.loc[row, "Cl"]
		root = df.loc[row, "Pāli Root"]
		root_in_comps = df.loc[row, "Root In Comps"]
		root_verb = df.loc[row, "V"]
		root_grp = df.loc[row, "Grp"]
		root_sign = df.loc[row, "Sgn"]
		root_meaning = df.loc[row, "Root Meaning"]
		base = df.loc[row, "Base"]
		family = df.loc[row, "Family"]
		construction =  df.loc[row, "Construction"]
		derivative = df.loc[row, "Derivative"]
		suffix = df.loc[row, "Suffix"]
		pc = df.loc[row, "Phonetic Changes"]
		comp = df.loc[row, "Compound"]
		comp_constr = df.loc[row, "Compound Construction"]
		source1 = df.loc[row, "Source1"]
		sutta1 = df.loc[row, "Sutta1"]
		eg1 = df.loc[row, "Example1"]
		source2 = df.loc[row, "Source 2"]
		sutta2 = df.loc[row, "Sutta2"]
		eg2 = df.loc[row, "Example 2"]
		ant = df.loc[row, "Antonyms"]
		syn = df.loc[row, "Synonyms – different word"]
		var = df.loc[row, "Variant – same constr or diff reading"]
		comm = df.loc[row, "Commentary"]
		notes = df.loc[row, "Notes"]
		cognate = df.loc[row, "Cognate"]
		category = df.loc[row, "Category"]
		metadata = df.loc[row, "Metadata"]
		link = df.loc[row, "Link"]

		colour1 = "#66FF99"
		colour2 = "#00A4CC"

		# css

		# headword
		
		tpp_js.write(f"""'{pali_clean}':'digital pāli dictionary<table><tr><td><span style="color:{colour2}">Pāli: </td><td>{pali}</td></tr>""")
		html_file.write(f"""<body><table><tr><td><span style="color:{colour2}">Pāli: </td><td>{pali}</td></tr>""")

		# meaning is empty > pos and buddhadatta
		
		if meaning == "":
			js_buddhadatta = re.sub("'", "\\'", buddhadatta)

			simple_js.write(f"""'{pali_clean}':'<b>{pali}</b>. {pos}. {js_buddhadatta} <span style="color:gray"><i>[under construction]</i></span>""")

			tpp_js.write(f"""<tr><td><span style="color:{colour2}">POS: </td><td>{pos}</td></tr>""")
			tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Meaning: </td><td><b>{js_buddhadatta}</b></td></tr></table>""")
			tpp_js.write(f"""<span style="color:gray"><i>word info under construction</i></span>""")

			html_file.write(f"""<tr><td><span style="color:{colour2}">POS: </td><td>{pos}</td></tr>""")
			html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Meaning: </td><td><b>{buddhadatta}</b></td></tr></table>""")
			html_file.write(f"""<span style="color:gray"><i>word info under construction</i></span>""")

		# meaning not empty

		if meaning != "":

			# grammar
			
			simple_js.write(f"""'{pali_clean}':'<b>{pali}</b>. {grammar}""")
			tpp_js.write(f"""<tr><td><span style="color:{colour2}">Grammar: </td><td>{grammar}""")
			html_file.write(f"""<tr><td><span style="color:{colour2}">Grammar: </td><td>{grammar}""")

			if neg != "":
				simple_js.write(f""", {neg}""")
				tpp_js.write(f""", {neg}""")
				html_file.write(f""", {neg}""")

			if verb != "":
				simple_js.write(f""", {verb}""")
				tpp_js.write(f""", {verb}""")
				html_file.write(f""", {verb}""")
			
			if trans != "":
				simple_js.write(f""", {trans}""")
				tpp_js.write(f""", {trans}""")
				html_file.write(f""", {trans}""")

			if case != "":
				simple_js.write(f" ({case})")
				tpp_js.write(f""" ({case})""")
				html_file.write(f""" ({case})""")
			
			tpp_js.write(f"""</td></tr>""")
			html_file.write(f"""</td></tr>""")

			# meaning
			js_meaning = re.sub("'", "\\'", meaning)

			simple_js.write(f""". {meaning}""")
			tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Meaning: </td><td><b>{js_meaning}</b>""")
			html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Meaning: </td><td><b>{meaning}</b>""")

			# lit
			js_lit = re.sub("'", "\\'", lit)

			if lit != "":
				tpp_js.write(f""", lit. {js_lit}""")
				html_file.write(f""", lit. {lit}""")

			tpp_js.write(f"""</td></tr>""")
			html_file.write(f"""</td></tr>""")

			# root

			if root != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Root: </td><td>{root} {root_grp} {root_sign} ({root_meaning})</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Root: </td><td>{root} {root_grp} {root_sign} ({root_meaning})</td></tr>""")

			# root in comps

			if root_in_comps != "" and root_in_comps != "0":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">√ in Comps: </td><td>{root_in_comps}</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">√ in Comps: </td><td>{root_in_comps}</td></tr>""")

			# base

			if base != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Base: </td><td>{base}</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Base: </td><td>{base}</td></tr>""")

			# construction
			
			if construction != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Construction: </td><td>{construction}</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Construction: </td><td>{construction}</td></tr>""")

			# derivative

			if derivative != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Derivative: </td><td>{derivative} ({suffix})</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Derivative: </td><td>{derivative} ({suffix})</td></tr>""")

			# phonetic
			js_pc = re.sub(" ", "", pc)
			js_pc = re.sub("'", "\\'", pc)

			if pc != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Phonetic: </td><td>{js_pc}</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Phonetic: </td><td>{pc}</td></tr>""")

			# compound & compound construction

			if comp != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Compound: </td><td>{comp} ({comp_constr})</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Compound: </td><td>{comp} ({comp_constr})</td></tr>""")

			# antonym

			if ant != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Antonym: </td><td>{ant}</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Antonym: </td><td>{ant}</td></tr>""")

			# synonym

			if syn != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Synonym: </td><td>{syn}</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Synonym: </td><td>{syn}</td></tr>""")

			# variant

			if var != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Variant: </td><td>{var}</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Variant: </td><td>{var}</td></tr>""")

			# commentary
			js_comm = re.sub("'", "\\'", comm)
			js_comm = re.sub("”", "\\'", js_comm)
			js_comm = re.sub("‘", "\\'", js_comm)
			js_comm = re.sub("’", "\\'", js_comm)

			if comm != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Commentary: </td><td>{js_comm}</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Commentary: </td><td>{comm}</td></tr>""")

			# notes
			js_notes = re.sub("'", "\\'", notes)
			js_notes = re.sub("”", "\\'", js_notes)
			js_notes = re.sub("“", "\\'", js_notes)
			js_notes = re.sub("‘", "\\'", js_notes)
			js_notes = re.sub("’", "\\'", js_notes)

			if notes != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Notes: </td><td>{js_notes}</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Notes: </td><td>{notes}</td></tr>""")
	
			# cognate

			if cognate != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Cognate: </td><td>{cognate}</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Cognate: </td><td>{cognate}</td></tr>""")

			# link

			if link != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Link: </td><td><a href="{link}">{link}</a></td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Link: </td><td><a href="{link}">{link}</a></td></tr>""")

			# non ia

			if non_ia != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Non IA: </td><td>{non_ia}</td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Non IA: </td><td>{non_ia}</td></tr>""")

			# sanskrit

			if sk != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Sanskrit: </td><td><i>{sk}</i></td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Sanskrit: </td><td><i>{sk}</i></td></tr>""")

			# sk root
			sk_root_mn = re.sub("'", "", sk_root_mn)	
		

			if sk_root != "":
				tpp_js.write(f"""<tr valign="top"><td><span style="color:{colour2}">Sk Root: </td><td><i>{sk_root} {sk_root_cl} ({sk_root_mn})</i></td></tr>""")
				html_file.write(f"""<tr valign="top"><td><span style="color:{colour2}">Sk Root: </td><td><i>{sk_root} {sk_root_cl} ({sk_root_mn})</i></td></tr>""")

			# end table

			tpp_js.write(f"""</table>""")
			html_file.write(f"""</table>""")

			# example 1
			js_eg1 = re.sub("'", "\\'", eg1)
			js_eg1 = re.sub("”", "\\'", js_eg1)
			js_eg1 = re.sub("“", "\\'", js_eg1)
			js_eg1 = re.sub("‘", "\\'", js_eg1)
			js_eg1 = re.sub("’", "\\'", js_eg1)
			js_eg1 = re.sub("{", "\\{", js_eg1)
			js_eg1 = re.sub("}", "\\}", js_eg1)

			if eg1 != "":
				tpp_js.write(f"""<br>{js_eg1}<br><span style="color:{colour2}"><i>{source1} {sutta1}</i></span>""")
				html_file.write(f"""<br>{eg1}<br><span style="color:{colour2}"><i>{source1} {sutta1}</i></span>""")

			# example 2
			js_eg2 = re.sub("'", "\\'", eg2)
			js_eg2 = re.sub("”", "\\'", js_eg2)
			js_eg2 = re.sub("“", "\\'", js_eg2)
			js_eg2 = re.sub("‘", "\\'", js_eg2)
			js_eg2 = re.sub("’", "\\'", js_eg2)
			js_eg2 = re.sub("{", "\\{", js_eg2)
			js_eg2 = re.sub("}", "\\}", js_eg2)
			js_sutta2 = re.sub("'", "\\'", sutta2)

			if eg2 != "":
				tpp_js.write(f"""<br><br>{js_eg2}<br><span style="color:{colour2}"><i>{source2} {js_sutta2}</i></span>""")
				html_file.write(f"""<br><br>{eg2}<br><span style="color:{colour2}"><i>{source2} {sutta2}</i></span>""")

		# new line
		
		if row < df_length-1:
			simple_js.write(f"""',\n""")
			tpp_js.write(f"""',\n""")
			html_file.write(f"""<br><br></body>""")

		if row == df_length-1:
			simple_js.write(f"""'\n""")
			tpp_js.write(f"""'\n""")
			html_file.write(f"""<br><br></body>""")

	# fin 
	simple_js.write("};")
	tpp_js.write("};")
	html_file.write(f"""</html>""")
```
